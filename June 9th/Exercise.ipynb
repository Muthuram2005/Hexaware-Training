```python
#80
activity_counts = rdd.countByValue()
most_common = max(activity_counts.items(), key=lambda x: x[1])
least_common = min(activity_counts.items(), key=lambda x: x[1])
total_activities = rdd.count()
unique_activities = rdd.distinct().count()

print("===== LEARNING PLATFORM USAGE SUMMARY =====")
print("Total Activities:", total_activities)
print("Unique Activities:", unique_activities)
print("Most Common Activity:", most_common[0], "-", most_common[1])
print("Least Common Activity:", least_common[0], "-", least_common[1])
print("\nActivity Frequencies:")
for activity, count in sorted(activity_counts.items()):
    print(activity, ":", count)
```
