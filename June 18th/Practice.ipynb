```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = (
    SparkSession.builder
    .appName("DeltaLakeColab")
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
    .config("spark.sql.catalog.spark_catalog",
            "org.apache.spark.sql.delta.catalog.DeltaCatalog")
)

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

 **View and Global Temp View**

```python
sales_data = [
    (1001, "Rahul", "Hyderabad", "Electronics", "Laptop", 65000),
    (1002, "Priya", "Bangalore", "Electronics", "Mobile", 25000),
    (1003, "Amit", "Mumbai", "Furniture", "Chair", 7000),
    (1004, "Sneha", "Delhi", "Furniture", "Table", 12000),
    (1005, "Farhan", "Chennai", "Fashion", "Shoes", 4500),
    (1006, "Neha", "Pune", "Fashion", "Watch", 8000),
    (1007, "Arjun", "Hyderabad", "Electronics", "TV", 45000),
    (1008, "Meera", "Bangalore", "Fashion", "Bag", 3000)
]

columns = [
    "order_id",
    "customer_name",
    "city",
    "category",
    "product",
    "amount"
]

sales_df = spark.createDataFrame(sales_data, columns)

display(sales_df.show())
```

```python
sales_df.createOrReplaceTempView("sales_temp")
```

```python
spark.sql("""
SELECT *
FROM sales_temp
""").show()
```

```python
result = spark.sql("""
SELECT
    city,
    SUM(amount) AS total_revenue
FROM sales_temp
GROUP BY city
ORDER BY total_revenue DESC
""")

result.show()
```

```python
sales_df.createOrReplaceGlobalTempView("sales_global")
```

```python
spark.sql("""
SELECT *
FROM global_temp.sales_global
""").show()
```
