```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

products = [
    (1, "Laptop", 800),
    (2, "Mouse", 20),
    (3, "Keyboard", 40),
    (4, "Monitor", 200)
]

df = spark.createDataFrame(products, ["product_id","product_name","price"])

display(df.show())

df.createOrReplaceTempView("products")

result = spark.sql("""
SELECT AVG(price) AS avg_price
FROM products
""")

display(result.show())
```
