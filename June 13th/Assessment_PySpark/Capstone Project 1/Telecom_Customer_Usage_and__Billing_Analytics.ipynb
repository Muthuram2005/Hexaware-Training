**Data Cleaning**

```python
#8
from pyspark.sql.functions import *

customer.filter(col("plan_id").isNull()).show()
```

```python
#9
usage.filter(col("data_used_gb").isNull()).show()
```

```python
#12
usage = usage.fillna({"data_used_gb":0})
usage.show()
```

```python
#13
payments = payments.fillna({"amount_paid":0})
payments.show()
```

```python
#14
payments = payments.fillna({"payment_mode":"Not Provided"})
payments.show()
```

```python
#15
customer = customer.fillna({"plan_id":"UNKNOWN"})
customer.show()
```

```python
#16
customer = customer.withColumn("data_quality_status",when(col("plan_id")=="UNKNOWN","Bad").otherwise("Good"))
customer.show()

usage = usage.withColumn("data_quality_status",when(col("data_used_gb")==0,"Bad").otherwise("Good"))
usage.show()

payments = payments.withColumn("data_quality_status",when((col("amount_paid")==0) | (col("payment_mode")=="Not Provided"),"Bad").otherwise("Good"))
payments.show()
```

**Transformations**

```python
#31
complete_df = complete_df.withColumn("usage_category",when(col("data_used_gb") >= 70, "Heavy User").when(col("data_used_gb") >= 30, "Medium User").otherwise("Low User"))
complete_df.show()
```

```python
#32
complete_df = complete_df.withColumn("payment_category", when(col("amount_paid") >= 1000, "High Payment").when(col("amount_paid") >= 500, "Medium Payment").otherwise("Low Payment"))
complete_df.show()
```

```python
#33
complete_df = complete_df.withColumn("churn_risk", when((col("status")=="Inactive") | (col("payment_status")!="Success"), "High Risk").when(col("data_used_gb") < 15, "Medium Risk").otherwise("Low Risk"))
complete_df.show()
```

```python
#34
complete_df = complete_df.withColumn("over_usage_gb",col("data_used_gb") - col("data_limit_gb"))
complete_df.show()
```

```python
#35
complete_df = complete_df.withColumn("over_usage_flag",when(col("over_usage_gb") > 0, "Yes").otherwise("No"))
complete_df.show()
```
