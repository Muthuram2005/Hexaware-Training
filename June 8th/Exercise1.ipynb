```python
from pyspark.sql import SparkSession
spark = SparkSession.builder\
.appName("PySpark Exercise") \
.getOrCreate()
spark
```

```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]

columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.select("employee_id","name").show()
```

```python
df.select("name","salary","city").show()
```

```python
df.select("employee_id", "name", "department", "salary").show()
```

```python
df.withColumn("salary_category", when(col("salary") >= 85000, "High")
.when(col("salary") >= 75000, "Medium")
.otherwise("Low")).show()
```

```python
df.withColumn("salary_category", when(col("salary") >= 85000, "High")
.when(col("salary") >= 75000, "Medium")
.otherwise("Low")).groupBy("salary_category").count().show()
```
