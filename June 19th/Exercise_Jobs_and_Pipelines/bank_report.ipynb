```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum

spark = SparkSession.builder.getOrCreate()

accounts = [
    (101, "Mumbai", 50000),
    (102, "Chennai", 30000),
    (103, "Mumbai", 20000),
    (104, "Bangalore", 40000)
]

df = spark.createDataFrame(
    accounts,
    ["account_id","city","balance"]
)

report = df.groupBy("city").agg(sum("balance").alias("total_balance"))

display(report.show())
```
