**CSV Dataset: flights.csv**

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = spark.read.csv("flights.csv", header=True, inferSchema=True)
df.show()
df.printSchema()
```

```python
#1
df.show()
```

```python
#2
df.select("passenger_name","from_city","to_city").show()
```

```python
#3
df.filter(df.status == "Confirmed").show()
```

```python
#4
df.filter(df.status == "Cancelled").show()
```

```python
#5
df.filter(df.status == "Pending").show()
```

```python
#6
df.filter(df.ticket_price > 20000).show()
```

```python
#7
df.filter(df.to_city.isin("Dubai","Singapore","London")).show()
```

```python
#8
df.count()
```

```python
#9
df.groupBy("airline").count().show()
```

```python
#10
df.groupBy("status").count().show()
```

```python
#11
from pyspark.sql.functions import *
df.filter(df.status == "Confirmed").agg(sum("ticket_price").alias("total_revenue")).show()
```

```python
#12
df.groupBy("airline").agg(avg("ticket_price").alias("average_ticket_price")).show()
```

```python
#13
df.agg(max("ticket_price").alias("highest_ticket_price")).show()
```

```python
#14
df.agg(min("ticket_price").alias("lowest_ticket_price")).show()
```

```python
#15
df.orderBy(df.ticket_price.desc()).show()
```

```python
#16
df = df.withColumn("tax",df.ticket_price*0.05)
df.show()
```

```python

#17
df = df.withColumn("final_price", df.ticket_price + df.tax)
df.show()
```

```python
#18
df= df.withColumn("price_category", when(df.ticket_price >= 30000, "Premium")
.when(df.ticket_price >= 10000, "Standard")
.otherwise("Budget"))
df.show()
```

```python

#19
df.groupBy("price_category").count().show()
```

```python
#20
df.filter(df.status == "Confirmed").write.mode("overwrite").parquet("confirmed_flights.parquet")
parquet_df = spark.read.parquet("confirmed_flights.parquet")
parquet_df.show()
```

**JSON Dataset: hotels.json**

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = spark.read.option("multiline","true").json("hotels.json")
df.show(truncate=False)
df.printSchema()
```

```python
#1
df.show(truncate=False)
```

```python
#2
df.select("hotel_name","city","rating").show()
```

```python
#3
df.filter(df.category == "Luxury").show(truncate=False)
```

```python
#4
df.filter(df.rating > 4.5).show(truncate=False)
```

```python
#5
df.filter(df.rooms_available > 15).show(truncate=False)
```

```python
#6
df.filter(df.price_per_night > 10000).show(truncate=False)
```

```python
#7
df.filter(df.city.isin("Dubai","London")).show(truncate=False)
```

```python
#8
df.filter(df.contact.phone.isNull()).show(truncate=False)
```

```python
#9
df.filter(df.contact.email.isNull()).show(truncate=False)
```

```python
#10
df.select("hotel_name","contact.phone","contact.email").show()
```

```python
#11
from pyspark.sql.functions import *
df.filter(array_contains(df.amenities,"wifi")).show(truncate=False)
```

```python
#12
df.filter(array_contains(df.amenities,"spa")).show(truncate=False)
```

```python
#13
df.groupBy("city").count().show()
```

```python
#14
df.groupBy("category").count().show()
```

```python
#15
df.groupBy("category").agg(avg("rating").alias("average_rating")).show()
```

```python
#16
df.groupBy("city").agg(avg("price_per_night").alias("average_price_per_night")).show()
```

```python
#17
df.agg(max("price_per_night").alias("highest_price_per_night")).show()
```

```python
#18
df = df.withColumn("total_potential_revenue",df.rooms_available * df.price_per_night)
df.show(truncate=False)
```

```python
#19
df.orderBy(df.rating.desc()).show(truncate=False)
```

```python
#20
hotel = df.select("amenities","category","city","contact","hotel_id","hotel_name","price_per_night","rating","rooms_available","total_potential_revenue")
hotel.write.mode("overwrite").parquet("hotels_flattened.parquet")
parquet_df = spark.read.parquet("hotels_flattened.parquet")
parquet_df.show(truncate=False)
```
