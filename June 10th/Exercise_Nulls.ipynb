```python
#18
df1 = df.withColumn("data_quality_status",when(col("city").isNull() | col("provider").isNull() | col("booking_amount").isNull() | col("booking_status").isNull() |
col("payment_mode").isNull(),"Incomplete").otherwise("Complete"))
df1.show()
```

```python
#19
df1.groupBy("data_quality_status").count().show()
```

```python
#22
df2 = df.na.fill({"booking_amount":0})
df2 = df2.withColumn("tax", df2.booking_amount * 0.05)
df2.show()
```
