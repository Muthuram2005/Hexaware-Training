**Window Functions**

```python
#61
w = Window.orderBy(desc("revenue"))
revenue_df = inner_df.groupBy("doctor_id","doctor_name","specialization","city").agg(sum("bill_amount").alias("revenue"))
revenue_df.withColumn("rank",rank().over(w)).show()
```

```python
#62
revenue_df.withColumn("dense_rank",dense_rank().over(w)).show()
```

```python
#63
revenue_df.withColumn("row_num",row_number().over(w)).show()
```

```python
#64
revenue_df.orderBy(desc("revenue")).show(1)
```

```python
#65
revenue_df.orderBy(desc("revenue")).show(3)
```

```python
#66
w1 = Window.partitionBy("specialization").orderBy(desc("revenue"))

revenue_df.withColumn("rank",rank().over(w1)).filter("rank=1").show()
```

```python
#67
revenue_df.withColumn("rank",rank().over(w1)).filter("rank<=2").show()
```

```python
#68
revenue_df.withColumn("running_revenue",sum("revenue").over(w.rowsBetween(Window.unboundedPreceding,Window.currentRow))).show()
```

```python
#69
revenue_df.withColumn("prev_revenue",lag("revenue").over(w)).show()
```

```python
#70
revenue_df.withColumn("next_revenue",lead("revenue").over(w)).show()
```

```python
#71
revenue_df.withColumn("previous",lag("revenue").over(w)).withColumn("difference",col("revenue")-col("previous")).show()
```

```python
#72
revenue_df.withColumn("next",lead("revenue").over(w)).withColumn("difference",col("revenue")-col("next")).show()
```

```python
#73
wc = Window.partitionBy("city").orderBy(desc("revenue"))

revenue_df.withColumn("rank",rank().over(wc)).filter("rank=1").show()
```
