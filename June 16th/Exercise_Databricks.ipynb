**Mini ETL Exercises**

```python
#91
source_df = spark.read.csv("sales.csv",header=True,inferSchema=True)
source_df.show()
```

```python
#92
etl_df = source_df.withColumn("total_amount",col("quantity")*col("price"))
etl_df.show()
```

```python
#93
completed_df = etl_df.filter(col("status")=="Completed")
completed_df.show()
```

```python
#94
revenue_df = completed_df.groupBy().agg(sum("total_amount").alias("total_revenue"))
revenue_df.show()
```

```python
#95
city_revenue = completed_df.groupBy("city").agg(sum("total_amount").alias("city_revenue"))
city_revenue.show()
```

```python
#96
category_revenue = completed_df.groupBy("category").agg(sum("total_amount").alias("category_revenue"))
category_revenue.show()
```

```python
#97
completed_df.write.mode("overwrite").parquet("/FileStore/sales_parquet")
```

```python
#98
sales_parquet = spark.read.parquet("/FileStore/sales_parquet")
sales_parquet.show()
```

```python
#99
sales_parquet.createOrReplaceTempView("sales_parquet_view")
spark.sql(" select * from sales_parquet_view").show()
```

```python
#100
spark.sql(""" select city,category,sum(total_amount) revenue from sales_parquet_view
group by city,category
order by revenue desc """).show()
```
